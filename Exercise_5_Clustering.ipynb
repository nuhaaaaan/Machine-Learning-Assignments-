### Step to do:

        1. Import the necessary packages from scikit-learn (KMeans, AgglomerativeClustering, and silhouette_score). (1 points)

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
```

        2. Read the Mall Customers dataset into the pandas DataFrame and display the last 10 samples. (1 points)

```python

mc_df = pd.read_csv('Mall_Customers.csv')

# Display the last 10 rows
mc_df.tail(10)
```

       3. Assign the two features Annual Income (k$)" and "Spending Score (1-100)" to X. (2 points)

```python
# Extract the relevant features
X = mc_df[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

        4. Visualize the the two featuerse ("Annual Income (k$)" And "Spending Score (1-100)) to understand the data structure. (2 points)

```python
# Visualize the data:
# Visualizing the two features to understand the data structure
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0], X[:, 1], c='blue', s=50)
plt.title('Mall Customers: Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

plt.figure(figsize=(10,6))
```

        6. implementation of k-means on the X features using K= 5. ( 2 points)

```python
# Applying K-means with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)

# Assigning clusters to each data point
y_kmeans = kmeans.predict(X)
```

        7. Plot the predicted clusters from k-means. ( 2 points)

```python
# Visualizing the predicted clusters from K-means
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.title('K-means Clustering (K=5) of Mall Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()


plt.figure(figsize=(10,6))
```

        8. compute and print the silhouette score of K-mean clustering method (2 point)

```python
# Compute silhouette score for K-means
silhouette_kmeans = silhouette_score(X, y_kmeans)
print(f'Silhouette Score for K-means: {silhouette_kmeans}')
```

        9. Implementation of Agglomerative Clustering on the X.

You will implement the Agglomerative Clustering three times using three types of linkages (single, complete and average)

#### a - single linkages:
* Implementation of Agglomerative Clustering with single linkages,  (1 point)
* Plot the predicted clusters. (1 point)
* compute the silhouette score  (1 point)

```python
# Implementation of Agglomerative Clustering with single linkages
# Agglomerative Clustering with single linkage
agg_single = AgglomerativeClustering(linkage='single', n_clusters=5)
y_single = agg_single.fit_predict(X)

```

```python
# Plot the predicted clusters

# Plotting the clusters
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0], X[:, 1], c=y_single, cmap='viridis', s=50)
plt.title('Agglomerative Clustering (Single Linkage) of Mall Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
plt.figure(figsize=(10,6))
```

```python
# Compute silhouette score for Single Linkage
silhouette_single = silhouette_score(X, y_single)
print(f'Silhouette Score for Single Linkage: {silhouette_single}')
```

#### b - complete linkages:
* Implementation of Agglomerative Clustering with complete linkages,  (1 point)
* Plot the predicted clusters. (1 point)
* compute the silhouette score  (1 point)

```python
# Implementation of Agglomerative Clustering with complete linkages
# Agglomerative Clustering with complete linkage
agg_complete = AgglomerativeClustering(linkage='complete', n_clusters=5)
y_complete = agg_complete.fit_predict(X)
```

```python
# Plot the predicted clusters

# Plotting the clusters
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0], X[:, 1], c=y_complete, cmap='viridis', s=50)
plt.title('Agglomerative Clustering (Complete Linkage) of Mall Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()


plt.figure(figsize=(10,6))

```

```python
# Compute silhouette score for Complete Linkage
silhouette_complete = silhouette_score(X, y_complete)
print(f'Silhouette Score for Complete Linkage: {silhouette_complete}')
```

#### Write your answer here

The silhouette score is a valuable metric that helps assess the quality of clustering by measuring how well each data point fits within its assigned cluster compared to other clusters. Higher silhouette scores typically indicate better-defined and well-separated clusters, while lower scores suggest overlapping or poorly defined clusters.

When comparing the clustering methods used in this analysisK-means and Agglomerative Clustering (with different linkage strategies)the silhouette scores provide a clear indication of their performance.

K-means Clustering: This method partitions the data into distinct, spherical clusters. If K-means shows a high silhouette score, it indicates that the algorithm has successfully grouped customers into well-separated clusters based on their annual income and spending score.

Agglomerative Clustering (Single Linkage): Single linkage tends to form clusters by merging the closest data points, often leading to long, chain-like clusters. If this method yields a lower silhouette score, it suggests that the clusters are not well-defined and may have poor separation.

Agglomerative Clustering (Complete Linkage): This method merges clusters by minimizing the maximum distance between points in the clusters. Complete linkage generally leads to more compact and well-separated clusters, which is often reflected in a higher silhouette score, indicating better performance in creating meaningful clusters.

Agglomerative Clustering (Average Linkage): Average linkage is a compromise between single and complete linkage, seeking a balance between compactness and separation. It may perform better than single linkage, but typically not as well as complete linkage in most cases.

Observations:
Silhouette Scores and Clustering Quality: A higher silhouette score generally indicates better-defined clusters. Therefore, clustering methods that produce higher scores, like K-means or Agglomerative Clustering with Complete Linkage, would be considered more effective for segmenting mall customers based on their income and spending behavior.

Performance of Different Methods: If K-means produces the highest silhouette score, it would suggest that it is the most suitable method for this task, as it clearly differentiates the groups. However, if Agglomerative Clustering with Complete Linkage provides a higher score, it would imply that this method's strategy of maximizing the inter-cluster distance is more effective for this dataset.

In conclusion, the silhouette scores provide a clear indication of how well each clustering method partitions the data, helping to identify the best approach for customer segmentation. By analyzing the silhouette scores, we can conclude which method produces the most meaningful and well-separated clusters, guiding us towards the most appropriate clustering strategy for this dataset.
